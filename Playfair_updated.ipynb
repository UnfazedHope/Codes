```python
def create_matrix(key):
    mat = []
    letter_added = []
    for i in key:
        if i not in letter_added:
            letter_added.append(i)
            
    for i in range(ord('A'),ord('Z')+1):
        if chr(i) == 'J':
            continue
        elif chr(i) in letter_added:
            continue
        else:
            letter_added.append(chr(i))
            
            
    for i in range(0,25,5):
        mat.append(letter_added[i:i+5])
    return mat

def generate_pairs(message):
    message = message.replace(' ','')
    pairs = list(message[:])
    i = 0
    
    while(i < len(pairs)-1):
        if pairs[i] == pairs[i+1]:
            pairs.insert(i+1,'X')
        i+=2
    if len(pairs)%2 !=0:
        pairs.append('X')

    final_pairs = []
    for i in range(len(pairs)//2):
        final_pairs.append([pairs[2*i],pairs[(2*i)+1]])
    return final_pairs

def indexof(k,mat):
    for i in range(5):
        for j in range(5):
            if mat[i][j] == k:
                return i,j
            
def playfair(key,msg,encrypt):
    mat = create_matrix(key)
    pairs = generate_pairs(msg)
    # (a,b)
    if encrypt:
        inc = 1
    else:
        inc = -1

    cipher_text = []
    for i in pairs:
        k1 = i[0]
        k2 = i[1]

        r1,c1 = indexof(k1,mat)
        r2,c2 = indexof(k2,mat)

        if r1 == r2:
            cipher_text += mat[r1][(c1+inc)%5] + mat[r2][(c2+inc)%5]

        elif c1 == c2:
            cipher_text += mat[(r1+inc)%5][c1] + mat[(r2+inc)%5][c2]
        else:
            cipher_text += mat[r1][c2] + mat[r2][c1]  
            
    return "".join(cipher_text)

key = 'Guptaaag'.upper()
msg = 'HelloBhavansh'.upper()

print(f'Key: {key}\n')
print(f'Message: {msg}\n')

mat = create_matrix(key)
pairs = generate_pairs(msg)

cipher_text = playfair(key,msg,True)
print(cipher_text)
decrypted_text = playfair(key,cipher_text,False)
print(decrypted_text)
```
